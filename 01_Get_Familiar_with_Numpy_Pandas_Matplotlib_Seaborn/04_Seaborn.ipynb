## Distribution plot
- distplot
- joinplot
- pairplot
- rugplot
- kdeplot

### [displot](https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot)

- It is the combination of histogram with the seaborn kdeplot() and rugplot() function.
- It shows the distribution of a [univariate set of observation](https://en.wikipedia.org/wiki/Univariate_(statistics)#:~:text=Univariate%20is%20a%20term%20commonly,salaries%20of%20workers%20in%20industry.).

```python
sns.distplot(tips['total_bill'])
```

```python
# We can remove the kdeplot
sns.distplot(tips['total_bill'], kde=False)
```

```python
# Adjust the number of bins
sns.distplot(tips['total_bill'], bins=50)
```

```python
# Plot the distribution on the vertical axis and change the color
sns.distplot(tips['total_bill'], vertical=True, color="#87a832")
```

### [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

- It used to explore relation ship between [bivariate data](https://en.wikipedia.org/wiki/Bivariate_data), as well as their distribution at the same time.

- There are several kind of plot to compare:
    - [scatter](https://chartio.com/learn/charts/what-is-a-scatter-plot/): use to observe relationship between 2 variable
    - [reg](https://seaborn.pydata.org/generated/seaborn.regplot.html): use to plot data and a linear regression model fit
    - [resid](https://www.statisticshowto.com/residual-plot/#:~:text=A%20residual%20plot%20is%20typically,that%20is%20non%2Dlinearly%20associated.): use to find problem with regression
    - [kde](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot): use for visualizing the probability density of a continuous variable
    - [hex](https://www.tutorialspoint.com/how-can-seaborn-library-be-used-to-display-a-hexbin-plot-in-python): when data is scattered unevently, it gets difficult to capture all the data point in a scatter plot then hexagonal binning comes into play.

```python
# scatter
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
# reg
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# resid
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
# kde
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# hex
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

### [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

- It show pairwise relationship in a dataset ( just for the numerical columns)
- Support a color hue argument ( for categorical columns )

```python
tips.head()
```

```python
# numerical columns
sns.pairplot(tips)
```

```python
# adding hue for categorical columns
sns.pairplot(tips, hue='sex')
```

```python
# change markers for categorical
sns.pairplot(tips, hue='smoker', markers =['s','o'])
```

```python
# force marginal to histogram or any kind of plot
sns.pairplot(tips, hue='sex', diag_kind='hist')
```

```python
# Force both diagonal and off-diagonal plotting style
sns.pairplot(tips,kind='kde')
```

### [rugplot](https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot)

- It is a plot of data for a single quantitative variable along an axis
- For simple you can imagine it is analogous to a histogram with zero-width bins.

```python
sns.rugplot(tips['total_bill'])
```

### [kdeplot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot)

- kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

## Categorical plot
- barplot & countplot
- boxplot & violinplot
- stripplot & swarmplot

### Barplot and countplot

**[Barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html)**
- It tell the relationship between the categorical and numerical value
- For more explaining detail, [check here](https://www.data-to-viz.com/graph/barplot)

```python
# By default the estimator is the mean
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# We can change the estimator to anything
# In here we want to know the standard deviation
sns.barplot(x='sex', y='total_bill',data= tips, estimator=np.std)
```

```python
# We can change the order
# We can useing hue to adding one more categorical
sns.barplot(x='total_bill', y='day', data=tips, hue='sex')
```

**[Count plot](https://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=countplot#seaborn.countplot)**
- The concept is very simple as it name, it display the counts of observations in each categorical bin using bars.

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='sex',data=tips, hue='day')
```

### Boxplot and Violinplot
- Both of this are used to show the distribution of categorical data.

**[Boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)**
-  shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. 
- The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# We can do this for entire dataframe
sns.boxplot(data=tips, palette='rainbow', orient='h') # set orient='h'
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

**[Violinplot](https://seaborn.pydata.org/generated/seaborn.violinplot.html)**
- It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. 
- Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

### Stripplot and Swarmplot

**[Stripplot](https://seaborn.pydata.org/generated/seaborn.stripplot.html)**
- The stripplot will draw a scatterplot where one variable is categorical. 
- A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)
```

**[Swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)**
- The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. 
- This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True)
```

### [Catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html)
- Catplot is the most general form of a categorical plot. It can take in a **kind** parameter to adjust the plot type:

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
```

### Combining categorical plot

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

### [Heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
- Heatmap is a graphical way to visualize visitor behavior data in the form of hot and cold spots employing a warm-to-cool color scheme. The *warm* colors indicated sections with the most visitor interaction, red being the area of highest interaction, and the *cool* colors indicate sections with the lowest interaction.
- In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you.

```python
tips.head()
```

```python
# Matrix form for correlation data
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

**For the flights data**

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```

### [Clustermap](https://seaborn.pydata.org/generated/seaborn.clustermap.html)
- Seaborns Clustermap function is great for making simple heatmaps and hierarchically-clustered heatmaps with dendrograms on both rows and/or columns.


```python
sns.clustermap(pvflights)
```

```python
# More options to get the information a little clearer like normalization
sns.clustermap(pvflights,cmap='coolwarm',standard_scale=1)
```
